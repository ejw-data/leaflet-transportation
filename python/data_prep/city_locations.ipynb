### Retrieve Data

```python
df = pd.read_excel('./data/lg-cities.xlsx')
df.head()
```

```python
df['long']=""
df['lat']=""
```

```python
for index, row in df.iterrows():
    city = row['City']
    state = row['State']
    country = 'United%20States%20of%20America'
    
    
    url_base = f'https://api.geoapify.com/v1/geocode/search?city={city}&state={state}&country={country}&format=json&apiKey={geoapify_key}'
    
    response = requests.get(url_base).json()
    
    try:
        df.loc[index, 'lat'] = response['results'][0]['lat']
        df.loc[index, 'long'] = response['results'][0]['lon']
        print(f'{city}, {state} is complete')
    except:
        print(f'{city}, {state} failed')
    
df.head()

```

```python
df.rename(columns={'April 1, 2020\nEstimates Base':'June2020'}, inplace=True)
```
